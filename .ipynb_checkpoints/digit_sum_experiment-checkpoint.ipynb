```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from class_random_array import *
from digit_sum import *
```

```python
my_array = random_array(1e6,min_max=(1,9999),ints=True)
```

```python
s = my_array.generate_normal()
```

```python
%%time
ds = []
for i in s:
    ds.append(digit_sum(i))
ds_array = np.array(ds)
```

```python
s_mu = np.mean(s)
s_sigma = np.std(s)
```

```python
s_count, s_bins, s_ign = plt.hist(s, bins=100, density=True)
plt.plot(s_bins,
         1/(s_sigma*np.sqrt(2*np.pi))*np.exp(-(s_bins-s_mu)**2/(2*s_sigma**2)),
         linewidth=2,
         color='r')
plt.show()
```

```python
ds_unique, ds_count = np.unique(ds_array, return_counts=True)
```

```python
plt.bar(ds_unique, ds_count)
plt.xticks(ds_unique)
plt.show()
```
