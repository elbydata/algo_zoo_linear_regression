```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from class_random_array import random_array
```

```python
my_array = random_array(10000,min_max=(1,999),ints=True)
```

```python
s = my_array.generate_normal()
```

```python
mu = np.mean(s)
sigma = np.std(s)
```

```python
count, bins, ign = plt.hist(s, 50, density=True)
plt.plot(bins,
         1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)),
         linewidth=2,
         color='r')
plt.show()
```
